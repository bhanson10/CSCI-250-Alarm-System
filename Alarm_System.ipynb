```python
def buzz():
    while True:
        pi.hardware_PWM(buzPin, 1000, int(0.25e6))
        time.sleep(1)
        pi.hardware_PWM(buzPin, 800, int(0.25e6))
        time.sleep(1)
```

```python
def buzzStop():
    GPIO.output(buzPin, False)
```

```python
def collectPhotoresistor():
    Vin =  3.3 # Volt
    R2  = 10.0 # kOhm
    Vou = adc.readADC(channel=0)
    Photoresistor = R2 * (Vin/Vou - 1)
    return Photoresistor
```
